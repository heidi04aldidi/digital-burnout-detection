```python
import numpy as np
import pandas as pd
import random
```

```python
NUM_SAMPLES = 1000

user_type = []
screen_time_hours = []
work_study_hours = []
sleep_hours = []
breaks_per_day = []
stress_level = []
mood = []
night_screen_usage = []
burnout_score = []
burnout_risk = []
```

```python
for i in range(NUM_SAMPLES):
    
    # 1. User type
    utype = random.choice(["Student", "Professional"])
    user_type.append(utype)
    
    # 2. Screen time
    if utype == "Student":
        screen = np.clip(np.random.normal(7.5, 2), 2, 14)
        work = np.clip(np.random.normal(6, 2), 2, 10)
        sleep = np.clip(np.random.normal(6.5, 1.5), 3, 9)
    else:
        screen = np.clip(np.random.normal(8.5, 1.8), 2, 14)
        work = np.clip(np.random.normal(9, 1.5), 4, 12)
        sleep = np.clip(np.random.normal(6, 1), 4, 8)
    
    screen_time_hours.append(round(screen, 1))
    work_study_hours.append(round(work, 1))
    sleep_hours.append(round(sleep, 1))
    
    # 3. Breaks
    breaks = np.clip(int(np.random.normal(3, 2)), 0, 10)
    breaks_per_day.append(breaks)
    
    # 4. Stress (linked to work + sleep)
    stress = int(
        np.clip(
            (work / 2) + (7 - sleep) + np.random.normal(0, 1),
            1,
            10
        )
    )
    stress_level.append(stress)
    
    # 5. Mood
    if stress >= 7:
        mood.append("Low")
    elif stress >= 4:
        mood.append("Neutral")
    else:
        mood.append("Good")
    
    # 6. Night screen usage
    night = "Yes" if (screen > 8 and sleep < 6.5) else random.choice(["Yes", "No"])
    night_screen_usage.append(night)
    
    # 7. Burnout score (internal)
    score = 0
    if screen > 9: score += 2
    elif screen > 7: score += 1
    
    if sleep < 6: score += 2
    elif sleep < 7: score += 1
    
    if stress >= 8: score += 2
    elif stress >= 6: score += 1
    
    if breaks <= 1: score += 2
    if night == "Yes": score += 1
    if mood[-1] == "Low": score += 1
    
    burnout_score.append(score)
    
    # 8. Burnout risk label
    if score <= 3:
        burnout_risk.append("Low")
    elif score <= 6:
        burnout_risk.append("Medium")
    else:
        burnout_risk.append("High")
```

```python
df = pd.DataFrame({
    "user_type": user_type,
    "screen_time_hours": screen_time_hours,
    "work_study_hours": work_study_hours,
    "sleep_hours": sleep_hours,
    "breaks_per_day": breaks_per_day,
    "stress_level": stress_level,
    "mood": mood,
    "night_screen_usage": night_screen_usage,
    "burnout_risk": burnout_risk
})
```
